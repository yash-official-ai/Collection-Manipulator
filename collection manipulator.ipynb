```python
students = []

print("===== Welcome to Student Data Organizer =====")

while True:
    print("\n1. Add Student")
    print("2. Display All Students")
    print("3. Update Student")
    print("4. Delete Student")
    print("5. Display Subjects")
    print("6. Exit")

    choice = input("Enter your choice: ")

    
    if choice == "1":

        name = input("Enter Name: ")
        age = int(input("Enter Age: "))
        grade = input("Enter Grade: ")

        subjects_input = input("Enter Subjects separated by comma: ")
        subjects = set(subjects_input.split(","))

        student_id = input("Enter Student ID: ")
        dob = input("Enter Date of Birth: ")

        
        student_info = (student_id, dob)

        
        student = {
            "info": student_info,
            "name": name,
            "age": age,
            "grade": grade,
            "subjects": subjects
        }

        
        students.append(student)

        print("Student Added Successfully!")

    
    elif choice == "2":

        if len(students) == 0:
            print("No student records found.")

        else:
            print("\n===== Student Records =====")

            for s in students:
                print(f"""
Student ID : {s['info'][0]}
Name       : {s['name']}
Age        : {s['age']}
Grade      : {s['grade']}
Subjects   : {", ".join(s['subjects'])}
DOB        : {s['info'][1]}
""")

    
    elif choice == "3":

        sid = input("Enter Student ID to update: ")

        found = False

        for s in students:

            if s["info"][0] == sid:

                new_age = int(input("Enter New Age: "))
                new_subjects = input("Enter New Subjects separated by comma: ")

                s["age"] = new_age
                s["subjects"] = set(new_subjects.split(","))

                print("Student Updated Successfully!")
                found = True
                break

        if found == False:
            print("Student Not Found!")

    
    elif choice == "4":

        sid = input("Enter Student ID to delete: ")

        found = False

        for i in range(len(students)):

            if students[i]["info"][0] == sid:

                del students[i]

                print("Student Deleted Successfully!")
                found = True
                break

        if found == False:
            print("Student Not Found!")

    
    elif choice == "5":

        all_subjects = set()

        for s in students:
            all_subjects.update(s["subjects"])

        print("Unique Subjects Offered:")
        print(all_subjects)

    
    elif choice == "6":

        print("Thank You for using Student Data Organizer!")
        break

    else:
        print("Invalid Choice!")
```
